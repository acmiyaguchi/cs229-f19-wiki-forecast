```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

mapping = pd.read_csv('../sample_data/trial_6/mapping.csv')
edges = pd.read_csv('../sample_data/trial_6/edges.csv')

g = nx.from_pandas_edgelist(
    edges, 
    source="src", 
    target="dst",
    edge_attr="weight",
    create_using=nx.DiGraph
)
```

```python
print(mapping.head())
print(edges.head())
print(nx.info(g))
```

```python
nx.draw(g)
```

```python
nx.draw_spectral(g)
```

```python
nx.density(g)
```

```python
nx.is_weakly_connected(g)
```

```python
def subgraph(g):
    print(nx.info(g))
    print(f"Density: {nx.density(g)}")
    print(f"Clustering: {nx.average_clustering(g)}")
    plt.clf()
    plt.figure(figsize=(16, 6))
    plt.subplot(121)
    nx.draw(g)
    plt.subplot(122)
    nx.draw_spectral(g)
    plt.show()

for k in [8, 16, 32, 64, 128, 256, 512]:
    subgraph(g.subgraph(mapping.id[:k]))
```

```python
gs = g.subgraph(mapping.id[:128])
nx.pagerank(gs, weight="weight")
```
