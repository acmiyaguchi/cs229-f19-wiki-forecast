## Page Visits

```python
page_visits = (
    spark.read
    .format("org.apache.spark.sql.cassandra")
    .options(table="page_visits", keyspace="wikipedia")
    .load()
)
```

```python
page_visits.printSchema()
page_visits.show()
```

```python
page_visits.count()
```
